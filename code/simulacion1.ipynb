```python
def simular_tanques(df_materiales, capacidad_tanques, dimensiones_tanques, tonelaje, densidad_pulpa, ge, intervalo, tiempo_total):
    """
    Simula el flujo de materiales en tanques asegurando que cada tanque transfiera su contenido 
    exactamente cada T_transferencia después de que el material llegue al tanque.
    """

    num_pasos = tiempo_total // intervalo
    num_tanques = len(capacidad_tanques)
    historial_tanques = defaultdict(list)

    # 🔹 Cálculo de volúmenes y porcentaje de sólidos
    porc_solidos = ((densidad_pulpa - 1000) / (((ge - 1) / ge) * densidad_pulpa)) * 100
    vol_liquido = (tonelaje * (100 - porc_solidos)) / porc_solidos
    vol_mineral = tonelaje / ge

    # 🔹 Calcular tiempos de transferencia para cada tanque
    T_transferencia = {
        t: round((math.pi * 0.25 * 24 * 60 * 0.85 * 0.02831 * (dimensiones_tanques[t][0] ** 2) * dimensiones_tanques[t][1] / (vol_liquido + vol_mineral)))
        for t in capacidad_tanques
    }
    print("Tiempos de transferencia calculados:", T_transferencia)

    # 🔹 Inicializar tanques vacíos y tiempo acumulado desde la última transferencia
    contenido = {t: {} for t in capacidad_tanques}
    tiempo_acumulado = {t: 0 for t in capacidad_tanques}  # Tiempo acumulado desde la última transferencia

    # Simulación
    for paso in range(num_pasos):
        tiempo_actual = paso * intervalo

        # 🔹 Entrada de material en el Tanque 1 cada intervalo de alimentación
        if paso < len(df_materiales):
            material_entrada = df_materiales.iloc[paso]["id_blending"]
            cantidad_entrada = df_materiales.iloc[paso]["ton_5min"]
            if material_entrada in contenido[1]:
                contenido[1][material_entrada] += cantidad_entrada
            else:
                contenido[1][material_entrada] = cantidad_entrada

        # 🔹 Transferencia entre tanques si el tiempo acumulado es mayor o igual al T_transferencia
        for t in range(1, num_tanques):  # Desde el Tanque 1 hasta el penúltimo
            if contenido[t]:  # Si hay material en el tanque actual
                tiempo_acumulado[t] += intervalo  # Acumular tiempo desde la última transferencia

                if tiempo_acumulado[t] >= T_transferencia[t]:  # Si es momento de transferir
                    # Mover el contenido del tanque actual al siguiente
                    if contenido[t + 1]:
                        # Si el tanque siguiente ya tiene material, sumar las cantidades
                        for material, cantidad in contenido[t].items():
                            if material in contenido[t + 1]:
                                contenido[t + 1][material] += cantidad
                            else:
                                contenido[t + 1][material] = cantidad
                    else:
                        # Si el tanque siguiente está vacío, copiar el contenido
                        contenido[t + 1] = contenido[t].copy()

                    # Vaciar el tanque actual después de la transferencia
                    contenido[t] = {}
                    # Reiniciar el tiempo acumulado para el tanque actual
                    tiempo_acumulado[t] = 0

        # 🔹 Guardar estado de los tanques en el historial
        for t in capacidad_tanques:
            historial_tanques[t].append({"Tiempo (min)": tiempo_actual, **contenido[t]})

    # Convertir historial en DataFrames
    df_resultados = {t: pd.DataFrame(historial) for t, historial in historial_tanques.items()}
    return df_resultados
```

```python
import pandas as pd
import math
from collections import defaultdict

def simular_tanques(df_materiales, capacidad_tanques, dimensiones_tanques, tonelaje, densidad_pulpa, ge, intervalo, tiempo_total):
    """
    Simula el flujo de materiales en tanques asegurando que cada tanque transfiera su contenido 
    exactamente cada T_transferencia después de que el material llegue al tanque.
    """

    num_pasos = tiempo_total // intervalo
    num_tanques = len(capacidad_tanques)
    historial_tanques = defaultdict(list)

    # 🔹 Cálculo de volúmenes y porcentaje de sólidos
    porc_solidos = ((densidad_pulpa - 1000) / (((ge - 1) / ge) * densidad_pulpa)) * 100
    vol_liquido = (tonelaje * (100 - porc_solidos)) / porc_solidos
    vol_mineral = tonelaje / ge

    # 🔹 Calcular tiempos de transferencia para cada tanque
    T_transferencia = {
        t: round((math.pi * 0.25 * 24 * 60 * 0.85 * 0.02831 * (dimensiones_tanques[t][0] ** 2) * dimensiones_tanques[t][1] / (vol_liquido + vol_mineral)))
        for t in capacidad_tanques
    }
    print("Tiempos de transferencia calculados:", T_transferencia)

    # 🔹 Inicializar tanques vacíos y tiempo acumulado desde la última transferencia
    contenido = {t: {} for t in capacidad_tanques}
    tiempo_acumulado = {t: 0 for t in capacidad_tanques}  # Tiempo acumulado desde la última transferencia

    # 🔹 Usar "hora_salida_molino" directamente como marca de tiempo
    df_materiales = df_materiales.copy()
    df_materiales["time"] = df_materiales["hora_salida_molino"]  # Mantener la fecha y hora original

    # Simulación
    for paso in range(num_pasos):
        if paso >= len(df_materiales):
            break  # Evitar errores si se supera el tamaño del DataFrame
        
        tiempo_actual = df_materiales.iloc[paso]["time"]  # Usar la fecha y hora directamente

        # 🔹 Entrada de material en el Tanque 1 cada intervalo de alimentación
        material_entrada = df_materiales.iloc[paso]["id_blending"]
        cantidad_entrada = df_materiales.iloc[paso]["ton_5min"]
        if material_entrada in contenido[1]:
            contenido[1][material_entrada] += cantidad_entrada
        else:
            contenido[1][material_entrada] = cantidad_entrada

        # 🔹 Transferencia entre tanques si el tiempo acumulado es mayor o igual al T_transferencia
        for t in range(1, num_tanques):  # Desde el Tanque 1 hasta el penúltimo
            if contenido[t]:  # Si hay material en el tanque actual
                tiempo_acumulado[t] += intervalo  # Acumular tiempo desde la última transferencia

                if tiempo_acumulado[t] >= T_transferencia[t]:  # Si es momento de transferir
                    # Mover el contenido del tanque actual al siguiente
                    if contenido[t + 1]:
                        # Si el tanque siguiente ya tiene material, sumar las cantidades
                        for material, cantidad in contenido[t].items():
                            if material in contenido[t + 1]:
                                contenido[t + 1][material] += cantidad
                            else:
                                contenido[t + 1][material] = cantidad
                    else:
                        # Si el tanque siguiente está vacío, copiar el contenido
                        contenido[t + 1] = contenido[t].copy()

                    # Vaciar el tanque actual después de la transferencia
                    contenido[t] = {}
                    # Reiniciar el tiempo acumulado para el tanque actual
                    tiempo_acumulado[t] = 0

        # 🔹 Guardar estado de los tanques en el historial con la fecha y hora exacta
        for t in capacidad_tanques:
            historial_tanques[t].append({"time": tiempo_actual, **contenido[t]})

    # Convertir historial en DataFrames
    df_resultados = {t: pd.DataFrame(historial) for t, historial in historial_tanques.items()}
    return df_resultados
```

```python
import pandas as pd
import numpy as np

class Cosecha:
    def __init__(self, df, fecha_cosecha_anterior, fecha_cosecha_actual):
        self.fecha_cosecha_anterior = pd.to_datetime(fecha_cosecha_anterior)
        self.fecha_cosecha_actual = pd.to_datetime(fecha_cosecha_actual)
        self.df = df
    
    def extraccion(self):
        mask_interval = (self.df["time"] >= self.fecha_cosecha_anterior) & (self.df["time"] <= self.fecha_cosecha_actual)
        df_interval = self.df.loc[mask_interval].copy()
        df_sin_time = df_interval.drop(columns=["time"])
        filas_vacias = df_sin_time.isna().all(axis=1)
        indices_vacios = filas_vacias[filas_vacias].index.tolist()  # Convertir a lista de índices
        indices_validos = [idx - 1 for idx in indices_vacios if idx - 1 in self.df.index]
        
        if indices_validos:  # Si hay índices válidos
            filas_anteriores = self.df.loc[indices_validos]
            filas_anteriores = filas_anteriores[filas_anteriores["time"] < self.fecha_cosecha_actual]
        else:
            filas_anteriores = pd.DataFrame()

        fecha_especifica = self.fecha_cosecha_actual
        fecha_menor = self.df[self.df["time"] <= fecha_especifica].iloc[-1] if not self.df[self.df["time"] <= fecha_especifica].empty else None
        fecha_mayor = self.df[self.df["time"] > fecha_especifica].iloc[0] if not self.df[self.df["time"] > fecha_especifica].empty else None
        
        if fecha_menor is not None and fecha_mayor is not None:
            distancia_menor = abs(fecha_menor["time"] - fecha_especifica)
            distancia_mayor = abs(fecha_mayor["time"] - fecha_especifica)
            
            if fecha_menor.drop("time").isna().all():
                fila_cercana = fecha_mayor.to_frame().T  # Si la fecha menor es NaN, tomar la mayor
            elif distancia_menor < distancia_mayor:
                fila_cercana = fecha_menor.to_frame().T
            elif distancia_mayor < distancia_menor:
                fila_cercana = fecha_mayor.to_frame().T
            else:
                fila_cercana = fecha_menor.to_frame().T  # Si están a la misma distancia, tomar la menor
        elif fecha_menor is not None:
            fila_cercana = fecha_menor.to_frame().T
        elif fecha_mayor is not None:
            fila_cercana = fecha_mayor.to_frame().T
        else:
            fila_cercana = pd.DataFrame()  # Si no hay fechas disponible
        df_resultado = pd.concat([filas_anteriores, fila_cercana])
        return df_resultado
```
